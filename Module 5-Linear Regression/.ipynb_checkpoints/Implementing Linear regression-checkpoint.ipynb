```python
data=pd.read_csv('Transformed_Housing_Data2.csv')
data.head()
#data already cleaned , transformed and imputed

##Basic preprocessing done
#STEP 1-Exploring the target and the independent variables.
#STEP 2-Treating the outliers and missing values in independent and target variables.
#STEP 3-Transforming the categorical variables into numerical variables using dummy encoding.
```

Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Y=data['Sale_Price']
X=scaler.fit_transform(data.drop(columns=['Sale_Price']))
X=pd.DataFrame(data=X,columns=data.drop(columns=['Sale_Price']).columns)
X.head()
```

```python
##check for the correlation bw independent variables , X.corr(),removing multicollinearity
X.corr()
```

```python
#pair of independent variable of correlation>0.5
k=X.corr()
#Result = [(element values) x (condition of final list) x (how to take different to store the final element)]
z=[[str(i),str(j)] for i in k.columns for j in k.columns if (k.loc[i,j]>abs(0.5))&(i!=j)]
z,len(z)
```

Treating multicollinearity using VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=X
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
##earlier we saw that there are 16 pairs of high correlation,multicollinearity,
#but when we calculated VIF its only 7 because VIF is calculated all the independent variables
```

Start
Calculate VIF of Variables
Any Variable
with VIF >5?
No
Stop

if Yes Remove Var
with highest VIF

then (Recalculate)


```python
def MC_remover(data):
    #calculating vif of all the variables in the data as we did before
    pd. Series
    vif=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
    if vif.max()>5:
        print(vif[vif== vif.max()].index[0],'has been removed')
        data = data.drop(columns = [vif[vif == vif.max()].index[0]])
        return data
    else:
        print('No Multicollinearity present anymore')
    return data
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
for i in range(7):
    vif_data=MC_remover(vif_data)
vif_data.head()
```

```python
#calculating vif for remaining columns
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF,len(vif_data.columns)
```

â€¢ Build linear regression model using Scikit-Learn.
â€¢ Make predictions for test dataset and evaluate
model performance on test dataset.

```python
x=vif_data
y=data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.3,random_state=1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#implementation of linear regression using sklearn librarywe dont have to define our own gradient descent
from sklearn.linear_model import LinearRegression
#this will create an instance of linear regression,which will normalize all the independent variables such that
#linear regression will no longer need to find the value of intercept
lr=LinearRegression(normalize=True)
#The Ir.fit() function implements the gradient descent
#and the complete procedure over the training data.
#gradient descent find the best set of parameters to model the data
lr.fit(x_train,y_train)
```

```python
lr.coef_
#values of m1,m2,m3,.... and value ofm0=0 as we have normalized the data
```

```python
predictions =lr.predict(x_test)
#This score() function calculates the RÂ² value of the model we have created.
lr.score(x_test,y_test)
```

```python
#8 steps performed
#verifying the assumptions of linear regression
#makingresidual plot
residuals = predictions- y_test
residual_table= pd.DataFrame({'residuals':residuals,'predictions':predictions})
#no independent variable was used while plotting mean regression modelhere we are using multiple independent variables
#the predicted values or Y^ values represent the effect/outcome of all variables combined
residual_table=residual_table.sort_values(by='predictions')
```

```python
z=[i for i in range(int(residual_table['predictions'].max()))]
k=[0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi=130,figsize=(17,7))
plt.scatter(residual_table['predictions'],residual_table['residuals'],color='red',s=2)
plt.plot(z,k,color='green',linewidth=3,label='regression line')
plt.ylim(-800000,800000)
plt.xlabel('fitted points(ordered by predictions)')
plt.ylabel('residuals')
plt.title('residual plot')
plt.legend()
plt.show()
```

```python
#2.distribution of errors
plt.figure(dpi=100,figsize=(10,7))
plt.hist(residual_table['residuals'],color='red',bins=200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('distribution of residuals')
plt.show()
#residuals follow the normal distribution
```

```python
#magnitude of coefficient using bargraph
#model coefficients
coefficients_table=pd.DataFrame({'column':x_train.columns,
                                'coefficients':lr.coef_})
coefficient_table=coefficients_table.sort_values(by='coefficients')
```

```python
plt.figure(figsize=(8,6),dpi=120)
x=coefficient_table['column']
y=coefficient_table['coefficients']
plt.barh(x,y)
plt.xlabel('coefficients')
plt.ylabel('variables')
plt.title('Normalized coefficient plot')
plt.show()
```
